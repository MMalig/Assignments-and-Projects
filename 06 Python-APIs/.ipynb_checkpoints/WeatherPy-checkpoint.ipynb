### Analysis
* Max Temperature reaches its highest point the closer the latitude is to 0
* The concentration of cities with a high Humidity % increases the further the latitude rises above 0
* Cloudiness and High Wind Speed can occurs across all latitudes
* There appear to be a large number of cities across all latitudes with no clouds

* You will need to add your API Key to the api_keys.py file

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    city = city.replace(" ", "&20")
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# URL for API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key

# Create lists for API data
name = []
lat = []
lng = []
temp_max = []
clouds = []
humidity = []
wind = []

# Record counter
counter = 1

# Print 
print(f"Beginning Data Retrieval")
print(f"-------------------------------")

#Loop through the cities list 
for city in cities:  
    
    # Append lists where city is found 
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        name.append(response["name"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        temp_max.append(response["main"]["temp_max"])
        clouds.append(response["clouds"]["all"])
        humidity.append(response["main"]["humidity"])
        wind.append(response["wind"]["speed"])
        record_name = response["name"]
        print(f"Processing Record {counter} of Set 1 | {record_name}")
        print(f"{url}&q={city}")
        
        # Count next record 
        counter = counter + 1
        
    # Print statement if city not found
    except:
        print("City not found. Skipping...")
    continue
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create a data frame using the appended lists
df = pd.DataFrame({"City Name": name,
                  "Lat": lat,
                  "Lng": lng,
                  "Max_Temp": temp_max,
                  "Cloudiness": clouds,
                  "Humidity": humidity,
                  "Wind_Speed": wind})

# Convert the data frame into a CSV File
df.to_csv(output_data_file)

# Display a summary of the data
df.info()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
#Create a scatter plot from the data frame
plt.scatter(df.Lat, df.Max_Temp, edgecolors="black", alpha="0.75")

#Add properties
plt.title("Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.grid()

#Save figure
plt.savefig("output_data/Max_Temp_vs_Lat.png")
```

```python
#Create a scatter plot from the data frame
plt.scatter(df.Lat, df.Humidity, edgecolors="black", alpha="0.75")

# Add properties
plt.title("Latitude vs. Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.grid()

# Save figure
plt.savefig("output_data/Humidity_vs_Lat.png")
```

```python
#Create a scatter plot from the data frame
plt.scatter(df.Lat, df.Cloudiness, edgecolors="black", alpha="0.75")

# Add properties
plt.title("Latitude vs. Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid()

# Save figure
plt.savefig("output_data/Cloudiness_vs_Lat.png")
```
